```python
import Orange
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from Orange.data.pandas_compat import table_from_frame, table_to_frame
from sklearn.metrics import r2_score
```

```python
model = pickle.load(open('model.pkcls', 'rb'))
#model
```

```python
table = Orange.data.Table.from_file("files/test.csv")
table
```

```python
y_pred = model(table)
print(y_pred.shape)
y_pred[:10]
```

```python
# r2
r2_score(table.Y, y_pred)
```

```python
plt.figure(figsize=(8, 8))
    
temp_df = pd.DataFrame({"y": table.Y, "y_pred": y_pred})
temp_df.plot.kde(figsize=(8, 8))
plt.show()
```
